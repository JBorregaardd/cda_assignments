# New dataframe

```python
import csv
import pandas as pd

data = pd.read_csv('data/case1Data.csv')

# One hot encode the categorical variable last 5 columns also one for missing values
data = pd.get_dummies(data, columns=data.columns[-5:], dummy_na=True)

# For each column, fill missing values with the mean of that column
for col in data.columns:
    data[col] = data[col].fillna(data[col].mean())

# Save df to csv
data.to_csv('data/case1Data_preprocessed.csv', index=False)
```
